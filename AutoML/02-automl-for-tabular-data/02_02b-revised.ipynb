```python
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.model_selection import train_test_split
```

```python
train_data = TabularDataset('../01-machine-learning-recap/data/train_data.csv')
```

```python
train_data.head()
```

```python
train, test = train_test_split(train_data, test_size=0.3, shuffle=True)
```

## Modeling 

```python
EVAL_METRIC = "f1"

# Trained models will be saved here
SAVE_PATH = "AutoGluonModels"

# Column to predict
LABEL = "Claim"
```

```python
predictor = TabularPredictor(
    label=LABEL,
    path=SAVE_PATH,
    eval_metric=EVAL_METRIC
)
```

```python
predictor = predictor.fit(train)
```

```python
# predictor.leaderboard(extra_info=True)
predictor.leaderboard()
```

```python
predictor.leaderboard(silent=True)
```

## Evaluation

```python
test_leaderboard = predictor.leaderboard()
```

```python
test_leaderboard = predictor.leaderboard(
    test, 
    silent=True,
    extra_info=True
)
```

```python
test_leaderboard
```

```python
y_pred_proba = predictor.predict_proba(test.drop(LABEL, axis=1))
```

```python
perf = predictor.evaluate_predictions(y_true=test[LABEL], y_pred=y_pred_proba)
```
